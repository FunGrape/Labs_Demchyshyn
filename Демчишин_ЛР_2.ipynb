```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()

# Завдання 1. Прочитати CSV файл

df = pd.read_csv("Global YouTube Statistics.csv" ,encoding="latin1")
print("Датасет успішно завантажено\n")

# Завдання 2. Вивести перші 5 рядків

print("Перші 5 рядків датасету:")
print(df.head(), "\n")

# Завдання 3. Вивести розміри датасету

print("Розмір датасету (рядки, стовпці):")
print(df.shape, "\n")

# Завдання 4. Перевірка пропусків

print("Кількість пропусків у кожному стовпці:")
print(df.isna().sum(), "\n")

# Замінюємо порожні значення на NaN
df = df.replace(" ", np.nan)

# Пробуємо перетворити числові колонки
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors="ignore")

# Завдання 5. Замінити пропуски на середні значення

df = df.fillna(df.mean(numeric_only=True))

# Змінюємо тип на float де можливо
df = df.astype(float, errors="ignore")

print("Пропуски замінені на середні значення\n")

# Завдання 6. Повторна перевірка пропусків

print("Кількість пропусків після обробки:")
print(df.isna().sum(), "\n")

# Завдання 7. Кількість унікальних країн

unique_countries = df["Country"].nunique()
print("Кількість унікальних країн:", unique_countries, "\n")

# Завдання 8. Діаграма розподілу переглядів

plt.figure()
df["video views"].plot(kind="hist", bins=30)

plt.title("Розподіл переглядів YouTube")
plt.xlabel("Кількість переглядів")
plt.ylabel("Частота")

plt.show()

# Завдання 9. Максимальні, мінімальні та середні перегляди

max_views = df["video views"].max()
min_views = df["video views"].min()
mean_views = df["video views"].mean()

print("Максимальна кількість переглядів:", max_views)
print("Мінімальна кількість переглядів:", min_views)
print("Середня кількість переглядів:", mean_views, "\n")

# Завдання 10. Країна з найбільшою кількістю завантажених відео

country_max_uploads = df.groupby("Country")["uploads"].sum().idxmax()

print("Країна з найбільшою кількістю завантажених відео:")
print(country_max_uploads, "\n")

# Завдання 11. Відео з найбільшою та найменшою кількістю uploads

max_upload_video = df.loc[df["uploads"].idxmax(), "title"]
min_upload_video = df.loc[df["uploads"].idxmin(), "title"]

print("Відео з найбільшою кількістю uploads:")
print(max_upload_video)

print("\nВідео з найменшою кількістю uploads:")
print(min_upload_video)
```

```python
# Аналіз датасету Amazon Top 50 Bestselling Books

import pandas as pd
import numpy as np
from google.colab import files
uploaded = files.upload()

# 1. Прочитати CSV файл

df = pd.read_csv("bestsellers with categories.csv", encoding="latin1")
print("Файл успішно завантажено\n")

# 2. Вивести перші 10 рядків

print("Перші 10 рядків датасету:")
print(df.head(10), "\n")

# 3. Розміри датасету

print("Розміри датасету (рядки, стовпці):")
print(df.shape, "\n")

# 4. Скільки книг у датасеті

print("Кількість книг у датасеті:")
print(df.shape[0], "\n")

# 5. Змінити назви колонок

df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
print("Нові назви колонок:")
print(df.columns, "\n")

# 6. Перевірка пропусків

print("Кількість пропусків у кожному стовпці:")
print(df.isna().sum(), "\n")

# Чи є пропуски
if df.isna().sum().sum() == 0:
    print("Пропусків у датасеті немає\n")
else:
    print("У датасеті є пропуски\n")

# 7. Унікальні жанри

genres = df["genre"].unique()
print("Унікальні жанри:")
print(genres, "\n")

# 8. Статистика цін

print("Максимальна ціна:", df["price"].max())
print("Мінімальна ціна:", df["price"].min())
print("Середня ціна:", df["price"].mean())
print("Медіанна ціна:", df["price"].median(), "\n")

# 9. Пошук і сортування

# Найвищий рейтинг
max_rating = df["user_rating"].max()
print("Найвищий рейтинг:", max_rating)

# Скільки книг мають цей рейтинг
print("Кількість книг з таким рейтингом:",
      df[df["user_rating"] == max_rating].shape[0], "\n")

# Книга з найбільшою кількістю відгуків
most_reviews_book = df.loc[df["reviews"].idxmax(), "name"]
print("Книга з найбільшою кількістю відгуків:")
print(most_reviews_book, "\n")

# Найдорожча книга 2010 року
book_2010_max_price = df[df["year"] == 2010].sort_values(by="price", ascending=False).iloc[0]["name"]
print("Найдорожча книга 2010 року:")
print(book_2010_max_price, "\n")

# Скільки Fiction книг у 2012
fiction_2012 = df[(df["genre"] == "Fiction") & (df["year"] == 2012)].shape[0]
print("Кількість Fiction книг у 2012:", fiction_2012, "\n")

# Книги з рейтингом 4.9 у 2010 або 2011
rating_books = df[(df["user_rating"] == 4.9) & (df["year"].isin([2010, 2011]))].shape[0]
print("Кількість книг з рейтингом 4.9 у 2010 або 2011:", rating_books, "\n")

# 10. Мінімальна і максимальна ціна за жанрами

price_stats = df[["genre", "price"]].groupby("genre").agg(["min", "max"])

print("Мінімальна і максимальна ціна за жанрами:")
print(price_stats)
```
