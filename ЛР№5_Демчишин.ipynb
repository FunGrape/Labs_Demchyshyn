```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt


# 1. Датасет Iris
df = px.data.iris()

# 1.1 3D-графік
fig1 = px.scatter_3d(
    df,
    x='sepal_length',
    y='sepal_width',
    z='petal_length',
    color='species',
    title='3D розподіл Iris'
)
fig1.update_layout(title_x=0.5)
fig1.show()

# 1.2 Гістограма
fig2 = px.histogram(
    df,
    x='petal_width',
    color='species',
    barmode='overlay',
    title='Розподіл petal_width'
)
fig2.update_layout(title_x=0.5)
fig2.show()

# 1.3 Лінійний графік
mean_df = df.groupby(['species', 'sepal_width'])['sepal_length'].mean().reset_index()

fig3 = px.line(
    mean_df,
    x='sepal_width',
    y='sepal_length',
    color='species',
    title='Середнє sepal_length залежно від sepal_width'
)
fig3.update_layout(title_x=0.5)
fig3.show()


# 2. Датасет Gapminder
df2 = px.data.gapminder()

# 2.1 Гістограма ВВП (2007, топ-10)
df_2007 = df2[df2.year == 2007].copy()
df_2007['gdp'] = df_2007['gdpPercap'] * df_2007['pop']

top10 = df_2007.nlargest(10, 'gdp')

plt.figure()
bars = plt.bar(top10['country'], top10['gdp'])

plt.title('Топ-10 країн за ВВП (2007)', loc='center')
plt.xticks(rotation=45)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval),
             ha='center', va='bottom')

plt.show()

# 2.2 Кругова діаграма (Європа, топ-5)
europe = df2[(df2.year == 2007) & (df2.continent == 'Europe')].copy()
europe['gdp'] = europe['gdpPercap'] * europe['pop']

top5 = europe.nlargest(5, 'gdp')

plt.figure()
plt.pie(top5['gdp'], labels=top5['country'], autopct='%1.1f%%')
plt.title('ТОП-5 країн Європи за ВВП', loc='center')
plt.show()

# 2.3 Стовпчаста діаграма (Океанія)
oceania = df2[df2.continent == 'Oceania']

pivot = oceania.pivot(index='year', columns='country', values='pop')

pivot.plot(kind='bar')

plt.title('Динаміка населення країн Океанії', loc='center')
plt.ylabel('Населення')
plt.xticks(rotation=45)

plt.show()


# 3. Тригонометричні функції

# 3.1 sin(x)
x = np.linspace(0, 2*np.pi, 100)
plt.figure()
plt.plot(x, np.sin(x))
plt.title('y = sin(x)', loc='center')
plt.grid()
plt.show()

# 3.2 cos(x)
plt.figure()
plt.plot(x, np.cos(x))
plt.title('y = cos(x)', loc='center')
plt.grid()
plt.show()

# 3.3 tan(x)
x = np.linspace(-np.pi/2 + 0.1, np.pi/2 - 0.1, 100)
plt.figure()
plt.plot(x, np.tan(x))
plt.title('y = tan(x)', loc='center')
plt.grid()
plt.show()
```
