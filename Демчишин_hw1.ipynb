```python
# 1. Створити одновимірний масив (вектор) з першими 20-ма натуральними числами та вивести його значення.
import numpy as np

arr1 = np.arange(1, 21)
print("Завдання 1")
print(arr1)

# 2. Створити двовимірний масив (матрицю) розміром 4x4, заповнити його одиницями та вивести його значення.
arr2 = np.ones((4, 4))
print("Завдання 2")
print(arr2)

# 3. Створити масив розміром 5x5, заповнити його випадковими дійсними числами в діапазоні від -1 до 1 та вивести його значення.
arr3 = np.random.uniform(-1, 1, (5, 5))
print("Завдання 3")
print(arr3)

# 4. Створити масив розміром 3x3, заповнити його випадковими дійсними числами в діапазоні від 0 до 1 та вивести його значення.
arr4 = np.random.rand(3, 3)
print("Завдання 4")
print(arr4)

# 5. Створити два одновимірних масиви розміром 6, заповнити їх випадковими цілими числами в діапазоні від 1 до 20 та виконати на них поелементні операції додавання, віднімання та множення.
a = np.random.randint(1, 21, 6)
b = np.random.randint(1, 21, 6)

print("Завдання 5")
print("a =", a)
print("b =", b)

print("Додавання:", a + b)
print("Віднімання:", a - b)
print("Множення:", a * b)

# 6. Створити дві матриці розміром 3x2 та 2x4, заповнити їх випадковими цілими числами в діапазоні від 1 до 5 та перемножити їх між собою.
m1 = np.random.randint(1, 6, (3, 2))
m2 = np.random.randint(1, 6, (2, 4))

print("Завдання 6")
print("m1 =\n", m1)
print("m2 =\n", m2)

result6 = np.dot(m1, m2)
print("Добуток =\n", result6)

# 7. Створити матрицю розміром 3x3, заповнити її випадковими дійсними числами в діапазоні від 0 до 1 та транспонувати її.
m3 = np.random.rand(3, 3)

print("Завдання 7")
print("Початкова матриця =\n", m3)
print("Транспонована матриця =\n", m3.T)

# 8. Створити дві матриці розміром 3x3, заповнити їх випадковими цілими числами в діапазоні від 1 до 10 та знайти їхню різницю.
m4 = np.random.randint(1, 11, (3, 3))
m5 = np.random.randint(1, 11, (3, 3))

print("Завдання 8")
print("m4 =\n", m4)
print("m5 =\n", m5)

print("Різниця =\n", m4 - m5)

# 9. Створити дві матриці розміром 3x3, заповнити їх випадковими цілими числами в діапазоні від -5 до 5 та знайти їх добуток.
m6 = np.random.randint(-5, 6, (3, 3))
m7 = np.random.randint(-5, 6, (3, 3))

print("Завдання 9")
print("m6 =\n", m6)
print("m7 =\n", m7)

print("Добуток =\n", np.dot(m6, m7))

# 10. Створити матрицю розміром 3x3, заповнити її випадковими дійсними числами в діапазоні від -1 до 1 та знайти вектор-стовпчик, що містить суму елементів кожного рядка матриці.
m8 = np.random.uniform(-1, 1, (3, 3))

print("Завдання 10")
print("Матриця =\n", m8)

row_sums = np.sum(m8, axis=1).reshape(-1, 1)
print("Вектор-стовпчик сум =\n", row_sums)
```
